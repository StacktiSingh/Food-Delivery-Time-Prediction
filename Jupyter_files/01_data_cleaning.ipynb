# 01_data_cleaning

```python
import pandas as pd
import numpy as np

# Load Dataset
df = pd.read_csv('../Datasets/train.csv')

# Basic Inspection
df.head()
df.info()
df.isnull().sum()
```

```python
# TARGET CLEANING

df['Time_taken(min)'] = [time[1] for time in df['Time_taken(min)'].str.split(' ')]
df['Time_taken(min)'] = df['Time_taken(min)'].astype(int)
```

```python
# WEATHER CLEANING
# =========================

df['Weatherconditions'] = df['Weatherconditions'].apply(lambda x: x.split(' ')[1])
```

```python
# DELIVERY PERSON AGE
# =========================

df['Delivery_person_Age'] = df['Delivery_person_Age'].replace('NaN ', np.nan)
df['Delivery_person_Age'] = pd.to_numeric(df['Delivery_person_Age'])
df['Delivery_person_Age'] = df['Delivery_person_Age'].fillna(df['Delivery_person_Age'].median())
df['Delivery_person_Age'] = df['Delivery_person_Age'].astype(int)
```

```python
# DELIVERY PERSON RATING
# =========================

df['Delivery_person_Ratings'] = df['Delivery_person_Ratings'].replace('NaN ', np.nan)
df['Delivery_person_Ratings'] = pd.to_numeric(df['Delivery_person_Ratings'])
df['Delivery_person_Ratings'] = df['Delivery_person_Ratings'].fillna(df['Delivery_person_Ratings'].median())
```

```python
# TIME COLUMNS CLEANING
# =========================

df['Order_Date'] = pd.to_datetime(df['Order_Date'])

df['Time_Orderd'] = df['Time_Orderd'].replace('NaN ', pd.NaT)
df['Time_Orderd'] = pd.to_datetime(df['Time_Orderd'], errors='coerce')
df = df.dropna(subset=['Time_Orderd'])

df['Time_Order_picked'] = pd.to_datetime(df['Time_Order_picked'], errors='coerce')
```

```python
# MULTIPLE DELIVERIES
# =========================

df['multiple_deliveries'] = df['multiple_deliveries'].replace('NaN ', np.nan)
df['multiple_deliveries'] = pd.to_numeric(df['multiple_deliveries'])
df['multiple_deliveries'] = df['multiple_deliveries'].fillna(df['multiple_deliveries'].mode()[0])
df['multiple_deliveries'] = df['multiple_deliveries'].astype(int)
```

```python
# FESTIVAL
# =========================

df['Festival'] = df['Festival'].replace('NaN ', np.nan)
df['Festival'] = df['Festival'].fillna(df['Festival'].mode()[0])
```

```python
# CITY
# =======================

df['City'] = df['City'].replace('NaN ', np.nan)
df['City'] = df['City'].fillna('Unknown')
```

```python
# âœ… Save cleaned dataset for next step
df.to_csv("cleaned_data.csv", index=False)
```
